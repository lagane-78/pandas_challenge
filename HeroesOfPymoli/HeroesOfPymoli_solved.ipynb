## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python


age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 1000]

age_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]



#binning by age
purchase_data["age_bin"] = pd.cut(purchase_data["Age"], age_bins, labels=age_labels)
purchase_data
age_df=pd.DataFrame(purchase_data.groupby('age_bin').SN.nunique())

#rename columns
age_df = age_df.rename(index=str, columns={'SN': 'Player Count'})

#add percentage columns
age_df['Percentage of Players'] = age_df['Player Count'] / sum(age_df['Player Count'])*100

#format new column with %
age_df = age_df.style.format({'Percentage of Players': '{:.2f}%'})
age_df


```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#binning by age
purchase_data["age_bin"] = pd.cut(purchase_data["Age"], age_bins, labels=age_labels, include_lowest=False)
purchase_groupby_age = purchase_data.groupby(["age_bin"])

purchase_groupby_age.head()

age_purchase_total = purchase_groupby_age.sum()["Price"].rename("Total Purchase Value")
age_average = purchase_groupby_age.mean()["Price"].rename("Average Purchase Price")
age_counts = purchase_groupby_age.count()["Price"].rename("Purchase Count")

avg_per_player = age_purchase_total/total_num_players*100

age_data = pd.DataFrame({"Purchase Count": age_counts, "Average Purchase Price": age_average, "Total Purchase Value": age_purchase_total, "Avg Total Purchase per Person": avg_per_player})


age_data["Average Purchase Price"] = age_data["Average Purchase Price"].map("${:,.2f}".format)
age_data["Total Purchase Value"] = age_data["Total Purchase Value"].map("${:,.2f}".format)
age_data ["Purchase Count"] = age_data["Purchase Count"].map("{:,}".format)
age_data["Avg Total Purchase per Person"] = age_data["Avg Total Purchase per Person"].map("${:,.2f}".format)
age_data = age_data.loc[:, ["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]

# Display table
age_data
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Spender = purchase_data.groupby("SN")

Top_Purchase_Count = Spender["Purchase ID"].count().round(2)
Avg_Spender_Price = Spender["Price"].mean().round(2)
Total_Spender_Purchase = Spender["Price"].sum().round(2)


Top_Spenders = pd.DataFrame({"Purchase Count": Top_Purchase_Count,
                               "Average Purchase Price" : Avg_Spender_Price,
                                "Total Purchase Value": Total_Spender_Purchase
})

Top_Spenders

# Formatting currency 
Top_Spenders = Top_Spenders.sort_values(["Total Purchase Value"], ascending=False).head(5) 
Top_Spenders["Average Purchase Price"] = Top_Spenders["Average Purchase Price"].map("${:.2f}".format)
Top_Spenders["Total Purchase Value"] = Top_Spenders["Total Purchase Value"].map("${:.2f}".format)
Top_Spenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Popular_Item = purchase_data.groupby(["Item ID", "Item Name"])

Popular_Purchase_Count = Popular_Item["Price"].count()
Popular_Purchase_Value = Popular_Item["Price"].sum()
Popular_Item_Price = Popular_Purchase_Value/Popular_Purchase_Count

Most_Popular_Items = pd.DataFrame({"Purchase Count" : Popular_Purchase_Count,
                                  "Item Price": Popular_Item_Price,
                                  "Total Purchase Value": Popular_Purchase_Value})

Most_Popular_Items = Most_Popular_Items.sort_values("Purchase Count", ascending=False)

# Currency/Formatting Added                                   
Most_Popular_Items["Item Price"] = Most_Popular_Items["Item Price"].map("${:,.2f}".format)
Most_Popular_Items["Total Purchase Value"] = Most_Popular_Items["Total Purchase Value"].map("${:,.2f}".format)
Most_Popular_Items.head(5)
```
